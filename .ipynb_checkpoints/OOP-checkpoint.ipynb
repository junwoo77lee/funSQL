```python
from datetime import datetime

class Note():
    # Self is just the instance of the thing
    def __init__(self, text, author):
        self.text = text
        self.author = author
        self.created_at = datetime.now()
        
    def save_me(self):
        # creating a behaviour
        # we're gonna save the data somehow to our machine
        with open(f"note_{self.author}_{self.created_at}.txt", "w") as f:
            f.write(f" {self.text} \n")
        
    def __str__(self):
         return f'The note text is {self.text}'
```

```python
note1 = Note("Take out the trash", "Jeff")
print(note1)
```

```python
print(note1.created_at)
```

```python
note1.save_me()
```

```python
# WARMUP
# I was asked this last week
# How can I get the median price of an item shown
# From a frequency table?
rows = [
    {"price": 50, "freq": 1900},
    {"price": 100, "freq": 800},
    {"price": 25, "freq": 250},
    {"price": 40, "freq": 750},    
]

# this means, items that were 50 bucks, were shown 1900 times
# what was the median value of "shown items?"
```

```python
temp = []

def cast_str_arry_to_int_arry(array):
    t = []
    for element in array:
        t.append(int(element))

    return t

for row in rows:
    s = str(row.get('price')) + ' '
    mult_val = int(row.get('freq'))
    temp += (cast_str_arry_to_int_arry((s * mult_val).strip().split(' ')))
    
# Use np.repeat()
# Use chain from itertools instead of list.extend
```

```python
len(temp)
```

```python
np.median(temp)
```
